```python
import numpy as np
#! Data Analysis and Statistics:
sales_data = np.array([300, 350, 250, 400, 200, 150, 250, 300, 350, 400])

avg_sales = np.mean(sales_data) 
max_sales = np.max(sales_data)
min_sales = np.min(sales_data)  

print("Average Sales : ", avg_sales)
print("Maximum Sales : ", max_sales)
print("Minimum Sales : ", min_sales)
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
from scipy.ndimage import convolve

#* Load image
img = Image.open('image.jpeg')
img_arr = np.array(img)
print(img_arr.shape)

#* Create a blur filter (3x3 averaging kernel)
kernel = np.ones((3, 3)) / 9.0

#* Apply filter
blurred_img = np.zeros_like(img_arr, dtype=float)
for c in range(3):  # for each RGB channel
    blurred_img[:, :, c] = convolve(img_arr[:, :, c], kernel, mode='reflect')

#* Convert back to uint8 and show
blurred_img = np.clip(blurred_img, 0, 255)  # Ensure values are in 0-255 range
blurred_img = blurred_img.astype(np.uint8)
# plt.imshow(img)
plt.imshow(blurred_img)
plt.axis('off')
plt.show()
```

```python

#! Machine Learning:

import numpy as np

#* Generate random input data
num_samples = 1000
input_data = np.random.randn(num_samples, 10)  # 1000 samples, 10 features

#* Generate random labels (binary classification)
labels = np.random.randint(0, 2, size=num_samples)  # 0 or 1

#* Example usage: Calculate mean and variance of features
mean_features = np.mean(input_data, axis=0)
var_features = np.var(input_data, axis=0)

print("Mean of features:", mean_features)
print("Variance of features:", var_features)
```

```python

#* Financial Modeling:
import numpy as np

#* Portfolio returns over time
returns = np.array([0.03, 0.02, -0.01, 0.05, -0.02])

#* Calculate cumulative returns
cumulative_returns = np.cumprod(1 + returns) - 1

print("Cumulative returns:", cumulative_returns)
```

```python

#* Signal Processing:
import numpy as np
import matplotlib.pyplot as plt

#* Generate a signal (sine wave)
t = np.linspace(0, 1, 1000)  # 1000 points from 0 to 1
signal = np.sin(2 * np.pi * 5 * t)  # 5 Hz sine wave

#* Add noise
noise = 0.2 * np.random.randn(1000)
noisy_signal = signal + noise

#* Plot the noisy signal
plt.plot(t, noisy_signal)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Noisy Signal')
plt.show()
```
