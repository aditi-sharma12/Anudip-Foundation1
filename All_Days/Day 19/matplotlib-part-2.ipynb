```python
import matplotlib.pyplot as plt
study_hours = [2, 3, 3, 4, 3, 5, 2, 6, 5, 7]
exam_scores = [65, 75, 60, 80, 70, 85, 70, 90, 88, 92]
# create a scatter plot
# plt.scatter(study_hours, exam_scores, c='green', marker='o', label='Student Data')
# plt.scatter(study_hours, exam_scores, c='green', marker='o', label='Student Data', s=exam_scores)
plt.scatter(study_hours, exam_scores, c='green', marker='o', label='Student Data', s=100, alpha=0.5, edgecolors='r', linewidths=10)
# Labeling and title 
plt.xlabel('Study Hours')
plt.ylabel('Exam Scores')
plt.title('study Hours vs. Exam Scores')
#show the scatter plot
plt.legend() 
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Sample student data
students= ['Chon', 'Smith', 'Marry', 'Rose', 'Devid'] 
math_scores = [85, 92, 78, 88, 90]
science_scores = [76, 88, 92, 80, 78]
# Create a subplot with the subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(6, 3))
# Plot Math Scores
axs[0].bar(students, math_scores, color='b')
axs[0].set_title("Math Scores")
axs[0].set_xlabel('Students') 
axs[0].set_ylabel('Score')
# Plot Science Scores
axs[1].bar(students, science_scores, color='g')
axs[1].set_title('Science Scores')
axs[1].set_xlabel('Students')
axs[1].set_ylabel('Score')
# Adjust spacing between subplats
plt.tight_layout()
# Show the subplot
plt.show()
```

```python
import matplotlib.pyplot as plt
#Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 5, 11]
#Creating plot
plt.plot(x, y)
#Adding title and axis Labels
plt.title('Plot with General Text')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
#Adding general text
plt.text(2, 8, 'General Text', fontsize=12, color='purple', bbox={"facecolor":"lightpink"}) 
plt.annotate(r'Java', xy=(4, 5), xytext=(3, 10), arrowprops=dict(facecolor='red', shrink = 0.1), fontsize=20, color='purple')
plt.legend(["Java"], loc=5, facecolor="lightblue", edgecolor="red", framealpha=0.5, shadow=True)
#Showing plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

# Create a plot
plt.figure(figsize=(8, 5))
plt.plot(x, y, marker='o')

# Set custom tick locations and labels for the x-axis
plt.xticks(ticks=[1, 2, 3, 4, 5], labels=['A', 'B', 'C', 'D', 'E'])

# Set custom tick locations and labels for the y-axis
plt.yticks(ticks=[10, 15, 20], labels=['Low', 'Medium', 'High'])

# Adding title and labels
plt.title('Custom Ticks Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Show plot
plt.grid(True)
plt.show()
```
