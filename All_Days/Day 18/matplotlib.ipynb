```python
# bar inside another bar

import numpy as np
import matplotlib.pyplot as plt
Categories = ['Category A', 'Category B', 'Category C', 'Category D']
values1 = [25, 40, 30, 35]
values2 = [20, 35, 25, 30]
bar_width = 0.35

n = len(Categories)

r1 = np.arange(n)
r2 = [x + bar_width for x in r1]

plt.figure(figsize = (10,6))
plt.bar(r1, values1, color='blue', width=bar_width, edgecolor='grey', label='Values 1')
plt.bar(r2, values2, color='green', width=bar_width, edgecolor='grey', label='Values 2')
plt.xlabel('Categories', fontsize=14)
plt.ylabel('Values', fontsize=14)
plt.title('Nested Bar Plot Example', fontsize=16)
# plt.xticks([r1 + bar_width/2 for r in range(n)], Categories)
plt.xticks(r1 + bar_width/2, Categories)  # Adjusted this line
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Corrected list of manufacturers
manufacturers = ['Samsung', 'Apple', 'Huawei', 'Xiaomi', 'Others']

# Market share percentages
market_share = [38, 25, 18, 12, 15]

# Colors for each slice
colors = ['lightblue', 'lightcoral', 'lightgreen', 'lightpink', 'lightgrey']

# Explode the last slice
ex = [0.12, 0, 0.05, 0.15, 0.1]
# Create a pie chart
plt.pie(market_share, labels=manufacturers, colors=colors, autopct='%1.1f%%', explode = ex, radius=0.9,
        shadow=True, labeldistance=1.1, counterclock=False,
        textprops={'fontsize':16, 'color':'b'},
        wedgeprops={'linewidth':4, 'edgecolor':'r'})

# Title
plt.title('Smartphone Market Share')
plt.legend(loc='upper left')
# Show the pie chart
plt.show()
```
