```python
import pandas as pd
import numpy as np

s = pd.Series(['A', 'B', 'C', 'A', 'B', np.nan, 'CABA', 'dog', 'cat'])

#  Basic Values Counts
print("Basic Values Counts :")
print(s.value_counts())

#  Do not sort by counts (order of appearance)
print("Basic Values Counts without sorting :")
print(s.value_counts(sort=False))

#  Sort by counts in ascending order
print("Basic Values Counts sorted in ascending order :")
print(s.value_counts(ascending=True))

#  Sort by counts in descending order
print("Basic Values Counts sorted in descending order :")
print(s.value_counts(ascending=False))

#  include nan values
print("Basic Values Counts including nan:")
print(s.value_counts(dropna=False))

print("Basic Values Counts excluding nan:")
print(s.value_counts(dropna=True))
s
```
